```python
#Let us take time spend on company as the dependent vairbale
# Since we have number_project aa the highest degree of corelation and last_evaluation as the second highest
```

```python
X=df.loc[:,['time_spend_company']]
y=df.number_project
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=40)
```

```python
X_train.shape,y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
MSE = mean_squared_error(y_test, y_pred)
MAE = mean_absolute_error(y_test, y_pred)
r2_score = r2_score(y_test, y_pred)
print(f"MSE = {MSE}")
print(f"MAE = {MAE}")
print(f"r2_score = {r2_score}")
```

# Let us check the error status by using both numberproject and lastevaluation as independent variable

```python
S=df.loc[:,["number_project","last_evaluation"]]

M=df.number_project
```

```python
from sklearn.model_selection import train_test_split
S_train, S_test, M_train, M_test = train_test_split(S, M, test_size=0.2, random_state=40)
model.fit(S_train,M_train)
```

```python
M_pred=model.predict(S_test)
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
MSE = mean_squared_error(M_test, M_pred)
MAE = mean_absolute_error(M_test, M_pred)
r2_score = r2_score(M_test, M_pred)
print(f"MSE = {MSE}")
print(f"MAE = {MAE}")
print(f"r2_score = {r2_score}")
```
